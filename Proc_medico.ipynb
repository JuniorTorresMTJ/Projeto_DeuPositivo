```python
cols = ['Unidade da Federação', '02 Procedimentos com finalidade diagnóstica',
       '03 Procedimentos clínicos', '04 Procedimentos cirúrgicos',
       '05 Transplantes de orgãos, tecidos e células']
```

```python
df_out = pd.read_csv("/content/Proc_Out.csv", encoding="ISO-8859-1", 
            skiprows=4,skipfooter = 12, sep=";", thousands=".", decimal =",", usecols =cols)
df_nov = pd.read_csv("/content/Proc_Nov.csv", encoding="ISO-8859-1", 
            skiprows=4,skipfooter = 12, sep=";", thousands=".", decimal =",", usecols =cols)
df_dez = pd.read_csv("/content/Proc_Dez.csv", encoding="ISO-8859-1", 
            skiprows=4,skipfooter = 12, sep=";", thousands=".", decimal =",", usecols =cols)
df_jan = pd.read_csv("/content/Proc_Jan.csv", encoding="ISO-8859-1", 
            skiprows=4,skipfooter = 12, sep=";", thousands=".", decimal =",", usecols =cols)
df_fev = pd.read_csv("/content/Proc_Fev.csv", encoding="ISO-8859-1", 
            skiprows=4,skipfooter = 12, sep=";", thousands=".", decimal =",", usecols =cols)
df_mar = pd.read_csv("/content/Proc_Mar.csv", encoding="ISO-8859-1", 
            skiprows=4,skipfooter = 12, sep=";", thousands=".", decimal =",", usecols =cols)
```

```python
s_out =pd.Series(['2020/Out', '2020/Out', '2020/Out', '2020/Out'])
s_nov =pd.Series(['2020/Nov', '2020/Nov', '2020/Nov', '2020/Nov'])
s_dez =pd.Series(['2020/Dez', '2020/Dez', '2020/Dez', '2020/Dez'])
s_jan =pd.Series(['2021/Jan', '2021/Jan', '2021/Jan', '2021/Jan'])
s_fev =pd.Series(['2021/Fev', '2021/Fev', '2021/Fev', '2021/Fev'])
s_mar =pd.Series(['2021/Mar', '2021/Mar', '2021/Mar', '2021/Mar'])
```

```python
df_out['mes'] = s_out
df_nov['mes'] = s_nov
df_dez['mes'] = s_dez
df_jan['mes'] = s_jan
df_fev['mes'] = s_fev
df_mar['mes'] = s_mar
```

```python
proc_medico  =  pd.concat([df_out, df_nov,df_dez, df_jan, df_fev, df_mar])
proc_medico 
```

```python
proc_medico['id_uf'] = proc_medico["Unidade da Federação"].str.slice(0,3)
proc_medico
```

```python
proc_medico["Unidade da Federação"] = proc_medico["Unidade da Federação"].str.slice(3,)
proc_medico.head()
```

```python
proc_medico.to_csv(index=False)
```
